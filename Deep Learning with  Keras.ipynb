### Deep Learning In Python WIth Keras


```python
# Load the EDA Packages
import pandas as pd
import numpy as np

# Load Visualization Packages
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load The Deep Learning Packages
from keras.models import Sequential
from keras.layers import Dense,Activation
```

### EDA

```python
names = [ ' preg ' , ' plas ' , ' pres ' , ' skin ' , ' test ' , ' mass ' , ' pedi ' , ' age ' , ' class ' ]
df = pd.read_csv("pima-indians-diabetes.data.csv",names=names)
```

```python
# Creating the Model
model = Sequential()
# input_dim = = no of features used as input data
# dense == used to define layers
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Check the Summary of the Layers built
model.summary()
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X = df.values[:,0:8]
```

```python
Y = df.values[:,8]
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# NB Epoch is the number of iterations
# Batch_size is the instances evaluated before weight is update

# Evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

#### Making Prediction

```python
# calculate predictions
predictions = model.predict(X)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
